## Assignment: Deal cards

```python
import requests
import json
```

### Shuffe the deck - get deck id

```python
# Initial url to shuffle the deck
url = "https://deckofcardsapi.com/api/deck/new/shuffle/?deck_count=1"
response = requests.get(url)

# Show the response from the API
print(json.dumps(response.json(), indent=2))
```

### Set Deck ID as variable to get cards

```python
deck_id = response.json()['deck_id']

deck_id
```

### Deal cards using the deck id

```python
# Get 5 cards from the deck
deal_cards = f"https://deckofcardsapi.com/api/deck/{deck_id}/draw/?count=5"

cards = requests.get(deal_cards)

# Show the response from the API
print(json.dumps(cards.json(), indent=2))
```

### Show value and suit of each card

Looping through the deck approach was consulted through the [Geeks For Geeks](https://www.geeksforgeeks.org/python/response-json-python-requests/) tutorial on `response.json()`.

```python
# Loop through the cards and print the value and suit of each card
for card in cards.json()['cards']:
    print(card['value'], card['suit'])
```
