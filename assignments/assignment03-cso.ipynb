## Assignment 3 - Retrieving Data From CSO

```python
import requests
import json
```

### Exchequer Account (Historical Series) - Dataset URL

```python
# Base URL for the dataset
url_beginning = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/"

# Ending URL for the dataset
url_end = "/JSON-stat/2.0/en"

# Table code (change for different datasets)
table_code = "FIQ02"
```

### Retrieve Data

```python
# Retrieve data from the API and return it as a JSON object
def get_data(table_code):
    response = requests.get(url_beginning + table_code + url_end)
    return response.json()
```

### Save Data as JSON File

```python
# Save the retrieved data into a JSON file
def data_into_file(table_code):
    with open("cso.json", "wt") as fp:
        json.dump(get_data(table_code), fp)
```

```python
# Execute the functions 
if __name__ == "__main__":
    data_into_file(table_code)
```
