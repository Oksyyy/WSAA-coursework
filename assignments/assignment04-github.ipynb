## Assignment 4

```python
import requests
import base64
from config import git_authkey as cfg
```

```python
api_url = "https://api.github.com/repos/Oksyyy/assignment_4/contents/assignment.txt"

api_key = cfg["git_authkey"]
```

### Access The File

```python
# Set up the headers for authentication
headers = {
        "Authorization": f"Bearer {api_key}",
        "Accept": "application/vnd.github+json"
        }
```

```python
# Get the current content of the file
response = requests.get(api_url, headers=headers)

print("GET status:", response.status_code)
```

```python
if response.status_code != 200:
    print("Error:", response.text)
    exit()
```

### File Information

```python
response_json = response.json()

response_json
```

```python
# Required for the PUT request to update the file
file_sha = response_json["sha"]
```

### Update File Content
Use `base64` module to decode base64 file content. Code adapted from [Geek For Geeks](https://www.geeksforgeeks.org/python/encoding-and-decoding-base64-strings-in-python/) tutorial.

```python
# Decode content
content_decoded = base64.b64decode(response_json["content"])
content_txt = content_decoded.decode("utf-8")

print(content_txt)
```

```python
# Peplace "Andrew" with "Oksana" in the file content
update_content = content_txt.replace("Andrew","Oksana")

print(update_content)
```

```python
# Encode updated content
update_enc = base64.b64encode(update_content.encode("utf-8")).decode("utf-8")

update_enc
```

### Push Updated File To The Repository

```python
# Create the payload for the PUT request
payload = {
    "message": "Updated file",
    "content": update_enc,
    "sha": file_sha
}
```

```python
# Commit update
put_response = requests.put(api_url, json=payload, headers=headers)

print("PUT status:", put_response.status_code)
print(put_response.json())
```

```python
if put_response.status_code in [200, 201]:
    print("File successfully updated.")
else:
    print("PUT failed:", put_response.text)
```
